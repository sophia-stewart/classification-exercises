```python
# function to read query result into dataframe
def get_telco_data():
    sql = '''
    SELECT *
    FROM customers
    JOIN contract_types USING(contract_type_id)
    JOIN internet_service_types USING(internet_service_type_id)
    JOIN payment_types USING(payment_type_id)
    '''
    telco = pd.read_sql(sql, get_url('telco_churn'))
    return telco
# check that function works
get_telco_data()
```

#### 4. Once you've got your `get_titanic_data`, `get_iris_data`, and `get_telco_data` functions written, now it's time to add caching to them. To do this, edit the beginning of the function to check for the local filename of `telco.csv`, `titanic.csv`, or `iris.csv`. If they exist, use the .csv file. If the file doesn't exist, then produce the SQL and pandas necessary to create a dataframe, then write the dataframe to a .csv file with the appropriate name.

```python
# edit functions to cache data and get data from .csv file once it has been cached
def get_titanic_data():
    # check for presence of .csv file
    if os.path.isfile('titanic.csv'):
        # if there, use .csv file to retrieve data
        return pd.read_csv('titanic.csv', index_col=0)
    else:
        # if no .csv file, read sql query results into dataframe, assign to variable
        titanic = pd.read_sql('SELECT * FROM passengers', get_url('titanic_db'))
        # create .csv file from dataframe
        titanic.to_csv('titanic.csv')
        return titanic

def get_iris_data():
    if os.path.isfile('iris.csv'):
        return pd.read_csv('iris.csv', index_col=0)
    else:
        iris = pd.read_sql('SELECT * FROM measurements JOIN species USING(species_id)', get_url('iris_db'))
        iris.to_csv('iris.csv')
        return iris

def get_telco_data():
    if os.path.isfile('telco.csv'):
        return pd.read_csv('telco.csv', index_col=0)
    else:
        sql = '''
        SELECT *
        FROM customers
        JOIN contract_types USING(contract_type_id)
        JOIN internet_service_types USING(internet_service_type_id)
        JOIN payment_types USING(payment_type_id)
        '''
        telco = pd.read_sql(sql, get_url('telco_churn'))
        telco.to_csv('telco.csv')
        return telco
```

```python
# combine previous steps into a single function
def clean_iris(iris):
    iris = iris.drop(columns = ['species_id', 'measurement_id'])
    iris.rename(columns={'species_name':'species'}, inplace=True)
    dummies = pd.get_dummies(iris['species'])
    iris = pd.concat([iris, dummies], axis=1)
    return iris
```

```python
# add split function, make one prep function to clean+split
from sklearn.model_selection import train_test_split
def split_iris(iris):
    train, test = train_test_split(iris, test_size=0.2, random_state=123, stratify=iris.species)
    train, validate = train_test_split(train, test_size=0.3, random_state=123, stratify=train.species)
    return train, validate, test
def prep_iris(iris):
    iris = clean_iris(iris)
    train, validate, test = split_iris(iris)
    return train, validate, test
```

- **For virginica & versicolor: Compare the mean petal_width using the Mann-Whitney test (scipy.stats.mannwhitneyu) to see if there is a significant difference between the two groups. Do the same for the other measurement types.**

```python
null_hypothesis = 'there is not a significant difference between the mean petal widths of virginica and versicolor.'
alt_hypothesis = 'there is a significant difference between the mean petal widths of virginica and versicolor.'
```

```python
from scipy import stats
alpha = 0.05
t, p = stats.mannwhitneyu(pw_virginica, pw_versicolor, alternative='two-sided')
if p < alpha:
    print('There is evidence supporting the rejection of the null hypothesis.')
    print('There is evidence supporting the alternative hypothesis that', alt_hypothesis)
else:
    print('We fail to reject the null hypothesis that',null_hypothesis)
    print('There is no evidence supporting the alternative hypothesis.')
```
